```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(20.0,10.0)
```

## reading data

```python
data=pd.read_csv('headbrain.csv')
print(data.shape)
```

```python
data.head()
```

```python
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
m=len(x)
numer=0
demon=0
for i in range(m):
    numer+=(x[i]-mean_x)*(y[i]-mean_y)
    demon+=(x[i]-mean_x)**2
b1=numer/demon
b0=mean_y-(b1*mean_x)
print(b1,b0)
```

```python
max_x=np.max(x)+100
min_x=np.min(x)-100
X=np.linspace(min_x,max_x,1000)
Y=b0+b1*x
plt.plot(X,Y,color='#58b970',label='Regression Line')
plt.plot(X,Y,color='#ef5423',label='scatter plot')
plt.xlabel('head')
plt.ylabel('brain')
plt.legend()
plt.show()
```
